```python
#Scatterplot
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data=df)
```

# Ã‡ARPRAZLAMALAR

```python
sns.scatterplot(x = "total_bill", y = "tip", hue="time", data = df);
```

```python
sns.scatterplot(x = "total_bill", hue = "time", y = "tip", data=df)
```

```python
sns.scatterplot(x = "total_bill", hue = "time",style = "time", y = "tip", data=df)
```

```python
sns.scatterplot(x = "total_bill", hue = "day",style = "day", y = "tip", data=df)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue="size", size = "size", data=df)
```

# Dogrusal iliskinin gosterilmesi

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = df) #lmplot lineer modelin kisaltilmisidir
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue="smoker", data = df)
```

```python
 sns.lmplot(x = "total_bill", y = "tip", hue="smoker", col="time", data = df) #buradaki col iki grafige bolmus oluyor
```

```python
 sns.lmplot(x = "total_bill", y = "tip", hue="smoker", col="time", row= "sex", data = df)
```
