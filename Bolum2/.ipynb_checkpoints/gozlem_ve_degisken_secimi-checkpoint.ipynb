# GOZLEM VE DEGISKEN SECIMI : loc and iLoc

```python
import numpy as np
```

```python
import pandas as pd
```

```python
m = np.random.randint(1,30, size=(10,3))
df = pd.DataFrame(m, columns=["var1","var2","var3"])
```

```python
df
```

```python
#loc - tanimlandigi sekliyle secim yapmak icin kullanilir.
```

```python
df.loc[0:3] #loc'da 3'u de almis oldu.
```

```python
#iloc - alisik olunan indexleme mantigiyla secim yapar.
```

```python
df.iloc[0:3] #burada ise 3'u almadan gosterir.
```

```python
#verilen indekslere sadik kalarak, tanimlandigi sekliyle secim yapmak icin loc kullaniliyor.
#iloc'da ise bilinen indeksleme mantigiyla secim yaptirmis oluyor. Klasik secim imkani kullanilir.
```

```python
df.iloc[0,0]
```

```python
df.iloc[:3,:2]
```

```python
df.loc[:3,"var3"]
```

```python
df.iloc[:3,1:3]
```

```python
df.iloc[0:3]["var3"]
```
