# JOIN ISLEMLERI

```python
import numpy as np
import pandas as pd
m = np.random.randint(1,30, size=(5,3))
df = pd.DataFrame(m, columns=["var1","var2","var3"])
```

```python
df
```

```python
df2 = df+99
```

```python
df2
```

```python
pd.concat([df,df2])
```

```python
pd.concat([df,df2],ignore_index= True) #ignore_index ile degiskenleri siralamis olduk. Az once sorunluydu.
```

```python
df2.columns=["var1","var2","degisken"]
df2
```

```python
pd.concat([df,df2], join = "inner") #kesisimli olanlari birlestirmis oldu.
```
