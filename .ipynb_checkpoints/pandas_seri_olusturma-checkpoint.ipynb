# PANDAS SERISI OLUSTURMAK

```python
import pandas as pd
```

* Bu bir pandas serisi olusturma komududur. Degerleri indexleriyle beraber tuttugu icin indexlerini de gosterir.

```python
pd.Series([1,2,3,4,5])
```

```python
seri = pd.Series([1,2,3,4,5])
type(seri)
```

```python
seri.axes #index bilgilerine erismis oluyoruz.
```

```python
seri.dtype
```

```python
seri.size #icindeki eleman sayisini verir.
```

```python
seri.ndim #boyutunu gosterir
```

```python
seri.values #degerlerine erismis oluyoruz.
```

```python
seri.head(3) #serinin ilk bes gosterimini getirir. Arguman olarak 3 dersek 3 tane getirir.
```

```python
seri.tail() #head'Ä±n tersidir, sondan bakmaktir.
```

```python
#INDEKS ISIMLENDIRMESI
```

```python
pd.Series([22,44,66,85,5], index=[1,3,5,6,7]) #indeks isimlendirmesini yapabiliyoruz.
```

```python
seri = pd.Series([22,44,66,85,5], index=["a","b","c","d","e"]) #goruldugu gibi bu formatta da indeks tanimlamasi yapabiliyoruz.
```

```python
seri["a":"c"]
```

```python
#SOZLUK UZERINDEN LISTE OLUSTURMAK
```

```python
sozluk = {"reg":10,"log":11, "cart":12}
```

```python
seri = pd.Series(sozluk)
```

```python
seri
```

```python
#iki seriyi birlestirerek seri olusturma
```

```python
pd.concat([seri,seri]) #seriyi tek bir array gibi dusunmemiz lazim. Bu sebeple bunlari yan yana ekleyemeyiz, alt alta eklenir.
```
